```python
import datetime
import webbrowser
import speech_recognition as sr
import pyttsx3
import sys

# Initialize the speech engine
engine = pyttsx3.init()

class myvar:
    m = 1

def speak(text):
    """Function to make the assistant speak."""
    engine.say(text)
    engine.runAndWait()

def greet_user():
    """Function to greet the user based on the time of day."""
    hour = datetime.datetime.now().hour
    if 0 <= hour < 12:
        greet = "Good morning!"
    elif 12 <= hour < 18:
        greet = "Good afternoon!"
    else:
        greet = "Good evening!"
    speak(greet)
    speak("I am tech studio assistant. How can I help you?")
    print(greet)

def take_command():
    """Function to take a command from the user via speech."""
    recognizer = sr.Recognizer()
    with sr.Microphone() as source:
        print("Listening...")
        audio = recognizer.listen(source)
        try:
            print("Recognizing...")
            command = recognizer.recognize_google(audio, language='en-in')
            print(f"Command received: {command}")
            return command.lower()
        except sr.UnknownValueError:
            speak("Sorry, I did not understand that.")
            return "none"
        except sr.RequestError:
            speak("Sorry, I am unable to reach the recognition service.")
            return "none"

def get_confirmation():
    """Function to get a yes or no confirmation from the user."""
    speak("Do you want to open the link? Please say Open or close.")
    command = take_command()
    return command

def open_website(command):
    """Function to open a website based on the command."""
    if 'open youtube' in command:
        webbrowser.open("https://youtube.com/@education_funchod_news?si=VvYa8Fhq-V4vCUP7")
        speak("Opening YouTube")
    elif 'open google' in command:
        webbrowser.open("https://google.com")
        speak("Opening Google")
    elif 'open github' in command:
        webbrowser.open('https://github.com/Prem7218')
        speak("Opening GitHub")
    elif 'open resume site' in command:
        webbrowser.open('https://pdfhost.io/v/XR3R5WK7K_3Prem_Radheshyam_Mahajan')
        speak("Opening Resume Site")
    elif 'open portfolio' in command:
        webbrowser.open('https://premradheshyammaha.wixstudio.io/portfolio')
        speak("Opening Portfolio")
    elif 'open enjoy game' in command:
        webbrowser.open('https://guessnumberproject.tiiny.site/index.html')
        speak("Opening Enjoy Game")
    elif 'open software engineer help' in command:
        webbrowser.open('https://chatgpt.com/g/g-sPpdQZSM6-software-engineer')
        speak("Opening Software Engineer Help")
    elif 'open linkedin' in command:
        webbrowser.open('https://www.linkedin.com/in/prem-radheshyam-mahajan/')
        speak("Opening LinkedIn")
    else :
        speak("Ok Bye...")
        sys.exit()

def tell_time():
    """Function to tell the current time."""
    time_str = datetime.datetime.now().strftime("%H:%M:%S")
    speak(f"The time is {time_str}")
    print(f"The time is {time_str}")

def handle_command(command):
    """Function to handle user commands."""
    if 'open youtube' in command or 'open google' in command or 'open github' in command or 'open resume site' in command or 'open portfolio' in command or 'open enjoy game' in command or 'open software engineer help' in command or 'open linkedin' in command:
        confirmation = get_confirmation()
        if 'open' in confirmation:
            open_website(command)
        else:
            speak("Thanks for coming!")
    elif 'thanks bye' in command or 'thanks for coming' in command:
        speak("You're welcome!")
        myvar.m = 0
    elif 'play music' in command:
        # Simulated playing music
        speak("Playing music (simulated)")
    elif 'the time' in command:
        tell_time()

if __name__ == '__main__':
    greet_user()
    while True:
        command = take_command()
        if myvar.m == 1:
            handle_command(command)
        else:
            break
```
